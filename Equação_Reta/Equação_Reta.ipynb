Altere os Coeficentes 

```python
# Coeficiente de inclinação
m = 0.9

# Coeficiente Linear
n = 0
```

```python
# Gerando valores para x
x = np.linspace(0,10,11);

# Calculando y com a equação da reta
y = m * x + n

# Plotando Gráfico
plt.grid()
plt.plot(x,y)
plt.xlim(-1, 10)
plt.ylim(-1, 10)
```
